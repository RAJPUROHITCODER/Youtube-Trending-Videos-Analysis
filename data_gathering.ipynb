```python
import requests
import pandas as pd
```

```python
API_KEY="AIzaSyAoVhwMR19mqkaca_dXharIhPT0hLOWoeI"
MAX_RESULTS=50
```

```python
country=requests.get(f"https://www.googleapis.com/youtube/v3/i18nRegions?part=snippet&key={API_KEY}")
item=country.json()["items"]
country_name=[]
for i in item:
    country_name.append(
        
        [i["id"],
        i["snippet"]["name"]]
    )
```

```python
language=requests.get(f"https://www.googleapis.com/youtube/v3/i18nLanguages?part=snippet&key={API_KEY}")
lang={}
for i in language.json()["items"]:
        lang[i["id"]]=i["snippet"]["name"]
```

```python
category=requests.get(f"https://www.googleapis.com/youtube/v3/videoCategories?part=snippet&regionCode=US&key={API_KEY}")
cat={}
for i in category.json()["items"]:
    cat[i["id"]]=i["snippet"]["title"]  
```

```python
dataset=[]
for j in range(len(country_name)):
    url=f"https://www.googleapis.com/youtube/v3/videos?part=snippet,statistics,contentDetails&chart=mostPopular&regionCode={country_name[j][0]}&maxResults={MAX_RESULTS}&key={API_KEY}"
    response=requests.get(url)
    data=response.json()["items"]
    for i in data:
        dataset.append({
            "id":i.get("id"),
            "Date-time":i["snippet"].get("publishedAt"),
            "duration":i["contentDetails"].get("duration"),
            "channel_id":i["snippet"].get("channelId"),
            "channel_title":i["snippet"].get("channelTitle"),
            "title":i["snippet"].get("title"),
            "category_id":cat[i["snippet"].get("categoryId")] if i["snippet"].get("categoryId") in cat else None,
            "live_broad_cast":i["snippet"].get("liveBroadcastContent"),
            "Audio":lang[i["snippet"].get("defaultAudioLanguage")] if i["snippet"].get("defaultAudioLanguage") in lang else None ,
            "view":i["statistics"].get("viewCount"),
            "like":i["statistics"].get("likeCount"),
            "commment":i["statistics"].get("commentCount"),
            "tag":i["snippet"].get("tags")[0] if type(i["snippet"].get("tags"))==list else None,
            "country":country_name[j][1],
            "country_iso":country_name[j][0],
            "link":f"https://www.youtube.com/watch?v={i.get('id')}"
             }
            )
    print(j)
df=pd.DataFrame(dataset)
df
```
